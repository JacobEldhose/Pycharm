```python
img = cv2.imread('D:\\Luminar\\DL\\Open-cv\\image.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(img,pt1=(400,300),pt2=(100,200),color=(255,255,0),thickness=cv2.FILLED)
cv2.circle(img,center=(350,250),radius=50,color=(255,51,255),thickness=cv2.FILLED)
cv2.putText(
    img,
    text="luminar",
    org=(150,250),
    fontFace=cv2.FONT_HERSHEY_DUPLEX,
    fontScale=1,
    color=(0,0,0),
    thickness= 2
)
plt.imshow(img)
```

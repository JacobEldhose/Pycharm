```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv')
df
```

```python
features = df.iloc[:,:-1]
target = df.iloc[:,-1]
```

```python
sns.regplot(x=features['R&D Spend'],y=target)
```

```python
sns.regplot(x=features['Marketing Spend'],y=target)
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_data = make_column_transformer((OneHotEncoder(handle_unknown="ignore"),['State']),remainder="passthrough")
```

```python
col_data
```

```python
features = col_data.fit_transform(features)
features
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
```

```python
df1 = pd.DataFrame({'Actual':y_test,"Predict":y_pred,"diff":y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
mean_absolute_percentage_error(y_test,y_pred)
```
